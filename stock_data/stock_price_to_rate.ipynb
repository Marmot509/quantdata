```python
import pandas as pd
import os

# read data from data/000001.csv to a dataframe
df = pd.read_csv(os.path.join('data', '000001.csv'), parse_dates=['trade_time'])

# find and open the csv file with 000001 in the file name from backward_adjust_factor folder, and read data to a dataframe
# the file names in the folder may contain other characters, like 000001.SZ.csv, so we need to find the file name with 000001 in it
backward_adjust_factor_file_name = [f for f in os.listdir('backward_adjust_factor') if '000001' in f][0]
baf_df = pd.read_csv(os.path.join('backward_adjust_factor', backward_adjust_factor_file_name), parse_dates=['date'])

#for each line other than the last line in baf_df, divide the hfq_factor by the next line's hfq_factor
#the last line's hfq_factor is 1, so we don't need to divide it
#save the result to a new column named 'hfq_factor_divided'
for i in range(len(baf_df) - 1):
    baf_df.loc[i, 'hfq_factor_divided'] = baf_df.loc[i, 'hfq_factor'] / baf_df.loc[i + 1, 'hfq_factor']



```
